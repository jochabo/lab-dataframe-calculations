## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
# Sub Problem 1
orders_amount_spent_uc = orders.groupby('CustomerID').agg({'amount_spent':sum}).reset_index()
orders_amount_spent_uc
```

```python
# Sub Problem 2
orders_amount_spent_uc_95 = orders_amount_spent_uc[orders_amount_spent_uc > np.percentile(orders_amount_spent_uc,95)]
orders_amount_spent_uc_95

VIP_customers = orders[orders['CustomerID'].isin(orders_amount_spent_uc_95['CustomerID'])]
VIP_customers # 37973 rows
```

```python
orders_amount_spent_uc_75_95 = orders_amount_spent_uc[(np.percentile(orders_amount_spent_uc,75)< orders_amount_spent_uc) & (orders_amount_spent_uc < np.percentile(orders_amount_spent_uc,95))]
orders_amount_spent_uc_75_95

preferred_customers = orders[orders['CustomerID'].isin(orders_amount_spent_uc_75_95['CustomerID'])]
preferred_customers # 148429 rows
```
